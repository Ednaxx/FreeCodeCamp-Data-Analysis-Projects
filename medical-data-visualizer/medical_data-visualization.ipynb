```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
print("Setup completed")
```

#### Import data

```python
df = pd.read_csv("medical_examination.csv", header=0)
df.head()
```

#### Add 'overweight' column

```python
def BMI(row):
    if (row.weight / (row.height / 100) ** 2 > 25):
        return 1
    else:
        return 0
df['overweight'] = df.apply(BMI, axis=1)
df
```

#### Normalize data by making 0 always good and 1 always bad. If the value of 'cholesterol' or 'gluc' is 1, make the value 0. If the value is more than 1, make the value 1.

```python
def normalizeCholesterol(row):
  if (row.cholesterol == 1):
    return 0
  else:
    return 1
    
def normalizeGluc(row):
  if (row.gluc == 1):
    return 0
  else:
    return 1

df.cholesterol = df.apply(normalizeCholesterol, axis=1)
df.gluc = df.apply(normalizeGluc, axis=1)
df
```

#### Draw Categorical Plot

#### Create DataFrame for cat plot using `pd.melt` using just the values from 'cholesterol', 'gluc', 'smoke', 'alco', 'active', and 'overweight'.

```python
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
# Group and reformat the data to split it by 'cardio'. Show the counts of each feature. You will have to rename one of the columns for the catplot to work correctly.
df_cat = df_cat.groupby(["cardio", "variable"]).value_counts().reset_index(name="total")
df_cat
```

#### Get the figure for the output

```python
fig = sns.catplot(
    data=df_cat, x="variable", y="total", col="cardio", hue="value",
    kind="bar"
).fig
```

#### Draw Heat Map

#### Clean the data

```python
df_heat = df[
(df['ap_lo'] <= df['ap_hi'])
& (df['height'] >= df['height'].quantile(0.025))
& (df['height'] <= df['height'].quantile(0.975))
& (df['weight'] >= df['weight'].quantile(0.025))
& (df['weight'] <= df['weight'].quantile(0.975))
]
df_heat
```

#### Calculate the correlation matrix

```python
corr = df_heat.corr()
corr
```

#### Draw the heatmap with 'sns.heatmap()'

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(16,6))
sns.heatmap(corr, mask=mask, square=True, annot=True, fmt=".1f", vmax=.3, center=0, linewidths=.5, cbar_kws={"shrink": .5})
```
